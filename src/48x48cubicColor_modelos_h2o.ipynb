```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import h2o
from h2o.automl import H2OAutoML
h2o.init()
from sklearn.metrics import accuracy_score
```

```python
subset = pd.read_pickle(join(PROJECT_PATH,'subset_Color48x48NormalizedFilter.pkl'))
```

```python
X = np.array(list(subset["image_array"]))
```

```python
X = X.reshape(X.shape[0], -1)
```

```python
X.shape
```

```python
type(X)
```

```python
train = h2o.H2OFrame(X.tolist())
```

```python
train.head(5)
```

```python
train.shape
```

```python
#subset["label"] = pd.Categorical(subset["label"], categories = ["MSI","MSS"], ordered = True)
#y = subset["label"].cat.codes
y = np.array(subset["label"])
y
```

```python
train_y = h2o.H2OFrame(y.tolist(), column_names=["label"])
```

```python
train_y.head(5)
```

```python
train_complete = train_y.cbind(train)
train_complete.describe()
```

```python
# Identify predictors and response
x = train_complete.columns
y = "label"
x.remove(y)
```

```python
train_complete[y] = train_complete[y].asfactor()
```

```python
train, test = train_complete.split_frame(ratios = [0.8], seed = 1)
```

```python
aml = H2OAutoML(max_models=20, seed=1, sort_metric='AUC')
aml.train(x=x, y=y, training_frame=train)
```

```python
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```

```python
test_set = test.drop("label")
y_true = test["label"]
```

```python
y_pred = aml.leader.predict(test_set)
```

```python
prediction = y_pred.as_data_frame()
```

```python
y_true_frame = y_true.as_data_frame()
```

```python
accuracy_score(y_true_frame["label"], prediction["predict"])
```
